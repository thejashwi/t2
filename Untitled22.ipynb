```python
def imput_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass ==1:
            return 37
        elif Pclass ==2:
            return 29
        else:
            return 24
        
```

```python
t2['Age']=t2[['Age','Pclass']].apply(imput_age,axis=1)
```
